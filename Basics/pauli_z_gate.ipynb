# Pauli Z Gate (Phase flip)

This notebook demonstrates the Pauli-Z gate, which flips the phase of a qubit without changing its measurement outcome.

##Objectives:

-Apply the Pauli-Z gate to a qubit
-measure the output
-Understand why the measurement result appears unchganged

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

#create a 1-Qubit circuit
qc = QuantumCircuit(1, 1)

#Apply Pauli-Z gate
qc.z(0)

#Measure the Qubit
qc.measure(0, 0)

#Run on simulator
backend = AerSimulator()
result = backend.run(qc, shots = 1024).result()

#get results
counts = result.get_counts()

#visualize results
plot_histogram(counts)
```

## Results and Explanations

The measurement result remains |0> with near 100% probability.

This is expected because the pauli-Z gate only flips the phase of the qubit.
Phase changes are not directly observable through measurements and become relevant when only interference happens.

## Remember this:

Measurement destroys interference because it eliminates superposition and erase relative phase by
converting amplitudes into classical probabilities.
