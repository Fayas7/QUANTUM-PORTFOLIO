# Two Qubit Quantum System

This notebook demonstrates how a 2-qubit quantum system behaves
and how joint measurement outcomes are observed.

## Objectives:

- create a 2-qubit quantum circuit
- measure both qubits
- observe joint measurement outcomes

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# No gates applied( both qubit starts in |0> state)
qc.measure([0,1], [0,1])

# Run on simulator
backend = AerSimulator()
result = backend.run(qc, shots =1024).result() 

counts = result.get_counts()
plot_histogram(counts)  
```

## Observations

The 2-qubit system collapses tot he joint state |00> upon measurement.
This shows that the measurement outcomes reptresent the combined state 
of both qubits, not independent single-bit results.
Why:
    - Both qubit starts in |0>.
    - No gates applied, which means deterministic output.
