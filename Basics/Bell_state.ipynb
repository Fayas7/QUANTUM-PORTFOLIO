# Bell State and Quantum Entanglement

This notebook demonstrates quantum entanglement using a Bell state
and shows how measurement outcomes are correlated.

## Objectives

- Create an entangled two-qubit Bell state
- Measure both qubits
- Observe correlated measurement outcomes

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# create a 2 qubit circuit
qc = QuantumCircuit(2, 2)

# step 1: apply hadamard gate to put first qubit into superposition (the control qubit).
qc.h(0)

# step 2: use CNOT gate to entangle the qubits.
qc.cx(0, 1)

# measure both qubits.
qc.measure([0,1], [0,1])

# Run on simulator
backend = AerSimulator()
result = backend.run(qc, shots = 1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

## Observations and Explanations:

The histogram shows only the outcomes of "00" and "11",
each occuring with approximately equal probability.

This demonstartes Quantum Entanglement: 
although each individual outcome is random, the results of
the 2 qubits are perfectly correlated.

## Additional points:

Entanglement does not allow faster-than-light communication.
Each outcome is still random.
Only the correlation is guaranteed.


Consider example:
if we measure:

| Alice (q0) | Bob (q1) |
| ---------- | -------- |
| 0          | 0        |
| 1          | 1        |

Alice’s result alone → random
Bob’s result alone → random
Relationship between them → perfectly correlated.
This correlastionis what entanglement preserves.
