# Pauli-X Gate (Quantum NOT)

This notebook demonstrates the pauli-X gate, which flips a qubit from |0> to |1>, similiar to classical NOT gate.

##Objectives:

-Apply Pauli-X gate to a single qubit.
-Measure the output.
-verify that |0> is flipped to |1>

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram 

#create 1 qubit circuit
qc = QuantumCircuit(1, 1)

#apply Pauli-X gate
qc.x(0)

#measure the qubit
qc.measure(0, 0)

#run on simulator
backend = AerSimulator()
result = backend.run(qc, shots=1024).result()

#get results
counts = result.get_counts()

#plot histogram
plot_histogram(counts)
```

##Result

The measurement shows that output state is |1> with near 100% probability.
This confirms that Pauli-X gate flips the quvbit grom |0> to |1>, acting as a quantum NOT gate.
