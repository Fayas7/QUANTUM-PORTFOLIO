# Hadamard Gate and Quantum Superposition

This notebook demonstrates how the Hadamard gate creates a superposition
state on a single qubit using Qiskit.


## Objective

- Apply a Hadamard gate to a single qubit
- Measure the qubit multiple times
- Observe equal probability of outcomes 0 and 1


```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Create a 1-qubit quantum circuit
qc = QuantumCircuit(1, 1)

# Apply Hadamard gate
qc.h(0)

# Measure the qubit
qc.measure(0, 0)

# Run on simulator
backend = AerSimulator()
result = backend.run(qc, shots=1024).result()

# Get measurement counts
counts = result.get_counts()

# Visualize results
plot_histogram(counts)
```

## Result

The histogram shows approximately equal counts for `0` and `1`.

This confirms that the Hadamard gate places the qubit into an equal
superposition of |0⟩ and |1⟩ before measurement.
