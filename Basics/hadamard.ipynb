1-Qubit Superposition using hadamard gate.

:this notebook demonstrates how a hadamard gate crates a quantum superposition on a single qubit using Qiskit.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)
qc
```

now code to run on simulator

```python
# Simulator
backend = AerSimulator()

# Run the circuit directly
result = backend.run(qc, shots=1024).result()
counts = result.get_counts()
```

Now to visualize cell

```python
plot_histogram(counts)
```

Observations:

After applying the hadamard gate , the qubit is placed in an equal superposition of |0⟩ and |1⟩.
Measurement results confirm this behavior with approximately equal probabilities.
